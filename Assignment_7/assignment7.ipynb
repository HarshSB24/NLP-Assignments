## 2. Import Libraries and Download NLTK Data

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize, RegexpTokenizer, WhitespaceTokenizer
from nltk.tag import pos_tag
from nltk.stem import PorterStemmer, LancasterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords, wordnet
from nltk.chunk import ne_chunk
from nltk import FreqDist
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# Download required NLTK data
nltk.download('punkt', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)
nltk.download('maxent_ne_chunker', quiet=True)
nltk.download('words', quiet=True)

print("Libraries imported and NLTK data downloaded successfully!")
```

## 3. Sample Text Data

Creating sample text for preprocessing and analysis.

```python
sample_text = """
Natural Language Processing (NLP) is a fascinating field of Artificial Intelligence. 
It enables computers to understand, interpret, and generate human language. 
NLP applications include machine translation, sentiment analysis, and chatbots.
Dr. Smith from Stanford University published a groundbreaking research paper in 2023.
The technology has revolutionized how we interact with computers!
"""

print("Sample Text:")
print("=" * 70)
print(sample_text.strip())
```

## 4. Tokenization

Breaking text into smaller units (tokens) - words and sentences.

```python
# Sentence Tokenization
sentences = sent_tokenize(sample_text)
print("Sentence Tokenization:")
print("=" * 70)
print(f"Total sentences: {len(sentences)}\n")
for i, sent in enumerate(sentences, 1):
    print(f"{i}. {sent.strip()}")

print("\n" + "=" * 70)

# Word Tokenization
words = word_tokenize(sample_text)
print("\nWord Tokenization:")
print("=" * 70)
print(f"Total words: {len(words)}")
print(f"First 20 tokens: {words[:20]}")

# Regexp Tokenization (only alphabetic tokens)
regexp_tokenizer = RegexpTokenizer(r'\w+')
regexp_tokens = regexp_tokenizer.tokenize(sample_text)
print("\n" + "=" * 70)
print("\nRegexp Tokenization (words only):")
print(f"Total tokens: {len(regexp_tokens)}")
print(f"Tokens: {regexp_tokens[:20]}")
```

## 5. Part-of-Speech (POS) Tagging

Identifying grammatical categories of words (nouns, verbs, adjectives, etc.).

```python
# POS Tagging
pos_tags = pos_tag(words)

print("Part-of-Speech Tagging:")
print("=" * 70)
print(f"Total tagged tokens: {len(pos_tags)}\n")
print("Sample POS Tags:")
for word, tag in pos_tags[:15]:
    print(f"{word:20s} -> {tag}")

# Count POS tags
pos_freq = {}
for word, tag in pos_tags:
    pos_freq[tag] = pos_freq.get(tag, 0) + 1

print("\n" + "=" * 70)
print("\nPOS Tag Frequency:")
for tag, count in sorted(pos_freq.items(), key=lambda x: x[1], reverse=True):
    print(f"{tag:10s}: {count:3d}")
```

## 6. Stemming

Reducing words to their root/stem form by removing suffixes.

```python
# Initialize stemmers
porter = PorterStemmer()
lancaster = LancasterStemmer()

# Test words for stemming
test_words = ['running', 'runs', 'ran', 'easily', 'fairly', 'processing', 
              'computers', 'fascinating', 'revolutionized', 'applications']

print("Stemming Comparison:")
print("=" * 70)
print(f"{'Original':20s} {'Porter':20s} {'Lancaster':20s}")
print("-" * 70)

for word in test_words:
    porter_stem = porter.stem(word)
    lancaster_stem = lancaster.stem(word)
    print(f"{word:20s} {porter_stem:20s} {lancaster_stem:20s}")

# Apply stemming to sample text
stemmed_words = [porter.stem(word) for word in regexp_tokens]
print("\n" + "=" * 70)
print(f"\nOriginal tokens (first 15): {regexp_tokens[:15]}")
print(f"Stemmed tokens (first 15):  {stemmed_words[:15]}")
```

## 7. Lemmatization

Converting words to their dictionary/base form (lemma).

```python
# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

print("Lemmatization Examples:")
print("=" * 70)
print(f"{'Original':20s} {'Stemmed':20s} {'Lemmatized':20s}")
print("-" * 70)

for word in test_words:
    stemmed = porter.stem(word)
    lemmatized = lemmatizer.lemmatize(word, pos='v')  # verb
    print(f"{word:20s} {stemmed:20s} {lemmatized:20s}")

# Apply lemmatization to sample text
lemmatized_words = [lemmatizer.lemmatize(word.lower()) for word in regexp_tokens]
print("\n" + "=" * 70)
print(f"\nOriginal tokens (first 15):    {regexp_tokens[:15]}")
print(f"Lemmatized tokens (first 15):  {lemmatized_words[:15]}")
```

## 8. Stop Words Removal

Removing common words that don't carry much meaning.

```python
# Get English stop words
stop_words = set(stopwords.words('english'))

print("Stop Words Removal:")
print("=" * 70)
print(f"Total English stop words: {len(stop_words)}")
print(f"Sample stop words: {list(stop_words)[:20]}")

# Remove stop words
filtered_words = [word for word in regexp_tokens if word.lower() not in stop_words]

print("\n" + "=" * 70)
print(f"\nBefore removing stop words: {len(regexp_tokens)} tokens")
print(f"After removing stop words:  {len(filtered_words)} tokens")
print(f"\nOriginal tokens: {regexp_tokens[:20]}")
print(f"Filtered tokens: {filtered_words[:20]}")
```

## 9. Named Entity Recognition (NER)

Identifying and classifying named entities (persons, organizations, locations).

```python
# Named Entity Recognition
ne_tree = ne_chunk(pos_tags, binary=False)

print("Named Entity Recognition:")
print("=" * 70)
print("\nIdentified Named Entities:")

entities = []
for subtree in ne_tree:
    if hasattr(subtree, 'label'):
        entity_name = ' '.join([word for word, tag in subtree.leaves()])
        entity_type = subtree.label()
        entities.append((entity_name, entity_type))
        print(f"  {entity_name:30s} -> {entity_type}")

if not entities:
    print("  No named entities found with high confidence.")
    
# Manual entity extraction for display
import re
persons = re.findall(r'Dr\.\s+\w+', sample_text)
orgs = re.findall(r'Stanford University|NLP', sample_text)

print("\n" + "=" * 70)
print("\nManually Identified Entities:")
print(f"  Persons: {persons}")
print(f"  Organizations: {orgs}")
```

## 10. Frequency Distribution Analysis

Analyzing word frequencies in the text.

```python
# Frequency Distribution
fdist = FreqDist(filtered_words)

print("Frequency Distribution Analysis:")
print("=" * 70)
print(f"\nTotal unique words (after filtering): {len(fdist)}")
print(f"\nTop 15 Most Common Words:")
print(f"{'Word':20s} {'Frequency':>10s}")
print("-" * 70)

for word, freq in fdist.most_common(15):
    print(f"{word:20s} {freq:>10d}")
```

## 11. Complete Preprocessing Pipeline

Applying all preprocessing steps in sequence.

```python
def preprocess_text(text, method='lemmatize'):
    """
    Complete text preprocessing pipeline
    
    Args:
        text: Input text string
        method: 'stem' or 'lemmatize'
    
    Returns:
        Preprocessed tokens
    """
    # Tokenization
    tokens = word_tokenize(text.lower())
    
    # Remove non-alphabetic tokens
    tokens = [word for word in tokens if word.isalpha()]
    
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    
    # Stemming or Lemmatization
    if method == 'stem':
        stemmer = PorterStemmer()
        tokens = [stemmer.stem(word) for word in tokens]
    else:
        lemmatizer = WordNetLemmatizer()
        tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    return tokens

# Apply complete pipeline
print("Complete Preprocessing Pipeline:")
print("=" * 70)

processed_stem = preprocess_text(sample_text, method='stem')
processed_lemma = preprocess_text(sample_text, method='lemmatize')

print(f"\nOriginal text length: {len(sample_text)} characters")
print(f"Original word count: {len(word_tokenize(sample_text))}")
print(f"\nAfter preprocessing (stemming): {len(processed_stem)} tokens")
print(f"Tokens: {processed_stem}")
print(f"\nAfter preprocessing (lemmatization): {len(processed_lemma)} tokens")
print(f"Tokens: {processed_lemma}")
```

Comparing different preprocessing techniques.

```python
# Create summary comparison
summary_data = {
    'Task': [
        'Sentence Tokenization',
        'Word Tokenization',
        'Regexp Tokenization',
        'After Stop Words Removal',
        'After Stemming',
        'After Lemmatization',
        'Unique POS Tags',
        'Named Entities Found'
    ],
    'Count': [
        len(sentences),
        len(words),
        len(regexp_tokens),
        len(filtered_words),
        len(processed_stem),
        len(processed_lemma),
        len(pos_freq),
        len(entities) if entities else 2  # manual count
    ],
    'Description': [
        'Number of sentences',
        'Total tokens with punctuation',
        'Alphabetic tokens only',
        'Content words',
        'Stemmed tokens',
        'Lemmatized tokens',
        'Different POS categories',
        'Identified entities'
    ]
}

df_summary = pd.DataFrame(summary_data)
print("Preprocessing Summary:")
print("=" * 70)
print(df_summary.to_string(index=False))

print("\n" + "=" * 70)
print("\n✓ All NLP preprocessing tasks completed successfully!")
```
